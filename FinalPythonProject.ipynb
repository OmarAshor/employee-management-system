```python
'''
اول حاجه محتاج انشاء ال
EmpolyeeManager Class
علشان هي الاساسيه و الاعتماد عليها و ال
Dictionary
علشان اخزن بيانات الموظفين فيها
'''
import csv # لازم استدعي مكتبه ال csv علشان هستخدمها

class EmployeeManager :

  #ال فاونكشن دي علشان افتح ديكشنري فاضي و علشان استخدم المكتبه بتاعت ال csv
  def __init__(self):
    self.employees = {} # dictionary علشان اخزن بيانات الموظفين
    self.load_from_csv() # الداله دي هتقراء من ملف ال csv و كل صف في الملف هيبقي عباره عن موظف في ال dictionary

  def load_from_csv(self):
        try: # using try & except because if the file is not found el brnameg myt3tlsh w ykml
            with open('employee.csv', 'r', newline="") as file: # esm el file + r means read only + don't add empty sottor
                reader = csv.DictReader(file)
                for row in reader:
                    emp_id = row["ID"] # using as a key
                    self.employees[emp_id] = {
                        "Name": row["Name"],
                        "Position": row["Position"],
                        "Salary": row["Salary"],
                        "Email": row["Email"]
                    }
        except FileNotFoundError:
            print("No existing data found. Starting with an empty employee list.")



  def add_employee(self):
        emp_id = input("Enter Employee ID: ")
        if emp_id in self.employees:
            print(f"Employee with ID {emp_id} already exists.")
            return

        name = input("Enter Employee Name: ")
        position = input("Enter Employee Position: ")
        salary = input("Enter Employee Salary: ")

        if not salary.isdigit():
            print("Invalid salary. Please enter a valid number.")
            return

        email = input("Enter employee Email: ").strip()
        valid_domains = [
            "@gmail.com", "@yahoo.com", "@outlook.com", "@hotmail.com", "@icloud.com",
            "@aol.com", "@msn.com", "@live.com", "@protonmail.com", "@mail.com",
            "@zoho.com", "@gmx.com", "@yandex.com"
        ]

        if not any(email.endswith(domain) for domain in valid_domains):
            print("Invalid email domain.")
            return

       #Save in the dictionary
        self.employees[emp_id] = {
            "Name": name,
            "Position": position,
            "Salary": salary,
            "Email": email
        }

        #Save in the csv
        with open('employee.csv', 'a', newline="") as file:
            writer = csv.DictWriter(file, fieldnames=["ID", "Name", "Position", "Salary", "Email"])
            if file.tell() == 0:
                writer.writeheader()

            writer.writerow({
                "ID": emp_id,
                "Name": name,
                "Position": position,
                "Salary": salary,
                "Email": email
            })

        print("Employee added successfully.")

  def view_all_employees(self):
        if not self.employees:
            print("No employees found.")
            return

        print("Employee List:")
        print("-" * 40)

        for emp_id, data in self.employees.items():
            print(f"ID: {emp_id}")
            print(f"Name: {data['Name']}")
            print(f"Position: {data['Position']}")
            print(f"Salary: {data['Salary']}")
            print(f"Email: {data['Email']}")
            print("-" * 40)

  def search_employee_by_id(self):
      emp_id = input("Enter Employee ID to search: ")  # اسأل المستخدم يدخل رقم الموظف

      if emp_id in self.employees:  # لو الرقم موجود في الديكشنري
          employee = self.employees[emp_id]  # هات بيانات الموظف من الديكشنري
          print("Employee Found:")
          print(f"ID: {emp_id}")
          print(f"Name: {employee['Name']}")
          print(f"Position: {employee['Position']}")
          print(f"Salary: {employee['Salary']}")
          print(f"Email: {employee['Email']}")
      else:
          print(f"No employee found with ID {emp_id}")


    # delete using id
  def delete_employee_by_id(self):
        emp_id = input("Enter Employee ID to delete: ")

        if emp_id in self.employees:
            confirm = input(f"Are you sure you want to delete employee {emp_id}? (yes/no): ")
            if confirm.lower() == 'yes':
                del self.employees[emp_id]  # delete from dictionary

                # اعيد كتابة ملف CSV من أول وجديد بعد الحذف
                with open('employee.csv', 'w', newline="") as file:
                    writer = csv.DictWriter(file, fieldnames=["ID", "Name", "Position", "Salary", "Email"])
                    writer.writeheader()  # اكتب الهيدر
                    for id, data in self.employees.items():
                        writer.writerow({
                            "ID": id,
                            "Name": data["Name"],
                            "Position": data["Position"],
                            "Salary": data["Salary"],
                            "Email": data["Email"]
                        })

                print(f"Employee with ID {emp_id} deleted successfully.")
            else:
                print("Deletion cancelled.")
        else:
            print(f"No employee found with ID {emp_id}")

    # update using id
  def update_employee_by_id(self):
        emp_id = input("Enter Employee ID to update: ")

        if emp_id not in self.employees:
            print(f"No employee found with ID {emp_id}")
            return

        employee = self.employees[emp_id]

        print("Leave input blank to keep current value.")

        name = input(f"Enter new name [{employee['Name']}]: ") or employee['Name']
        position = input(f"Enter new position [{employee['Position']}]: ") or employee['Position']
        salary = input(f"Enter new salary [{employee['Salary']}]: ") or employee['Salary']
        email = input(f"Enter new email [{employee['Email']}]: ").strip() or employee['Email']

        if not salary.isdigit():
            print("Invalid salary. Please enter a number.")
            return

        valid_domains = [
            "@gmail.com", "@yahoo.com", "@outlook.com", "@hotmail.com", "@icloud.com",
            "@aol.com", "@msn.com", "@live.com", "@protonmail.com", "@mail.com",
            "@zoho.com", "@gmx.com", "@yandex.com"
        ]

        if not any(email.endswith(domain) for domain in valid_domains):
            print("Invalid email domain.")
            return

        # for save in dictionary
        self.employees[emp_id] = {
            "Name": name,
            "Position": position,
            "Salary": salary,
            "Email": email
        }

        # for save in csv
        with open('employee.csv', 'w', newline="") as file:
            writer = csv.DictWriter(file, fieldnames=["ID", "Name", "Position", "Salary", "Email"])
            writer.writeheader()
            for id, data in self.employees.items():
                writer.writerow({
                    "ID": id,
                    "Name": data["Name"],
                    "Position": data["Position"],
                    "Salary": data["Salary"],
                    "Email": data["Email"]
                })

        print(f"Employee with ID {emp_id} updated successfully.")


# Main Menu
if __name__ == "__main__":
    manager = EmployeeManager()  # create instnace of the class

    while True:
        print("\n========= Employee Management System =========")
        print("1. Add New Employee")
        print("2. View All Employees")
        print("3. Search Employee by ID")
        print("4. Delete Employee by ID")
        print("5. Update Employee by ID")
        print("6. Exit")
        print("==============================================")

        choice = input("Enter your choice (1-6): ")

        if choice == '1':
            manager.add_employee()
        elif choice == '2':
            manager.view_all_employees()
        elif choice == '3':
            manager.search_employee_by_id()
        elif choice == '4':
            manager.delete_employee_by_id()
        elif choice == '5':
            manager.update_employee_by_id()
        elif choice == '6':
            print("Exiting program. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter a number from 1 to 6.")

```
